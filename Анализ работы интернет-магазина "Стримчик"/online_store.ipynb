<h1>Сборный проект-1. Анализ работы интернет-магазина "Стримчик"<span class="tocSkip"></span></h1>

По данным о продажах игр, оценках пользователей и экспертов, жанрах и платформах (например, Xbox или PlayStation) нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Описание данных
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

Необходимо ответить на вопросы:
- сколько игр выпускалось в разные годы
- за какой характерный срок появляются новые и исчезают старые платформы
- как менялись продажи по платформам
- найти лидирующие по продажам платформы 
- как влияют отзывы пользователей и критиков на продажи внутри одной популярной платформы, а затем внутри всех популярных

Для пользователя каждого региона определить:
- самые популярные платформы
- самые популярные жанры
- влияет ли рейтинг ESRB на продажи в отдельном регионе

Проверить гипотезы:
- равны ли средние пользовательские рейтинги платформ Xbox One и PC
- равны ли средние пользовательские рейтинги жанров Action и Sports

## Шаг 2. Подготовка данных

### Переименуем столбцы

```python
for name in df.columns:
    df = df.rename(columns={name:name.lower()})
df.info()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep='first')
```

Удалили 2 неявных дубликата: оставили строки, где больше заполненных данных.

### Заполняем пропуски и меняем типы данных

```python
round(df.isna().mean()*100,2).reset_index()
```

```python
df = df.dropna(subset=['year_of_release'])
```

Процент пропусков в столбце 'year_of_release' мал, поэтому можем удалить строки с пропущенными данными в нем.

```python
df['year_of_release'] = df['year_of_release'].astype(int)
df.info()
```

```python
df.loc[df['name'].isna()]
```

```python
df = df.dropna(subset=['name'])
df.info()
```

```python
df['user_score'].value_counts()
```

```python
df.loc[df['user_score'] == 'tbd', 'user_score'] = -1
```

```python
df['user_score'] = pd.to_numeric(df['user_score'])
df.info()
```

```python
df['rating'].value_counts()
```

```python
df.loc[df['rating'] == 'K-A', 'rating'] = 'E'
df['rating'].value_counts()
```

```python
df['rating'] = df['rating'].fillna('-')
```

```python
def type_rating(r):
    if r == 'E':
        return '6+'
    elif r == 'T':
        return '13+'
    elif r == 'M':
        return '17+'
    elif r == 'E10+':
        return '10+'
    elif r == 'EC':
        return '3+'
    elif r == 'RP':
        return 'демо'
    elif r == 'AO':
        return '18+'
    elif r == '-':
        return 'отсутствует'

df['type_rating'] = df['rating'].apply(type_rating)
df
```

### Считаем  продажи по регионам

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df
```

Мы изменили регистр во всех названиях столбцов на нижний. 

В столбце с годом выпуска релиза пропусков менее 2%, поэтому удалили строки с пропущенными датами и изменили тип данных на 'int', т.к. указана не полная дата, а только год.

Увидели, что в столбце с именем есть 2 пропущенных значения, в этих же строках пропущены практически все значения. Удалим эти 2 строки. Возможно эти пропуски появились при некорректной выгрузке данных.

В столбце 'user_score' тип строка и есть значения 'tbd' (будет определено). Заменили эти значения на '-1' и поменяли тип на 'float'

В интернете нашли расшифровку возрастной категории от организации ESRB:
- EC - 3+
- E - 6+
- E10+ - 10+
- T - 13+
- M - 17+
- AO - 18+
- RP - демо-версии
- K-A - 6+ (ранее было К-А, а сейчас Е, поэтому изменили на Е)

и добавили новый столбец с расшифровкой.

Организация ESRB была организована в 1994 году, а данные по играм у нас с 1980 г, поэтому есть пропуски в столбце 'rating'. Заполним эти пропуски '-' и укажем в графе тип рейтинга - 'отсутствует'

Пропуски в столбцах с рейтингами критиков и пользователей заполнять не будем, т.к. у нас нет достоверной информации. Пропуски в рейтингах могут быть связаны с тем, что какие-то игры не так популярны/распространены

### Исследуем влияние отзывов критиков и пользователей на продажи внутри платформы PS4.

```python
df_actuality.query('platform == "PS4"')['critic_score'].corr(df_actuality.query('platform == "PS4"')['total_sales']).round(2)
```

```python
df_actuality.query('platform == "PS4" & user_score != -1')['user_score'].corr(df_actuality.query('platform == "PS4"')
                                                                              ['total_sales']).round(2)
```

Как видим, корреляция не очень большая. Возможно это по той причине, что продажи не сильно зависят от отзывов. Скорее всего больше зависит от рекламы, уникальности игры.

## Проверка гипотез

### Исследуем пользовательские рейтинги платформ Xbox One и PC

Нулевая гипотеза: средние пользовательские рейтинги платформ XOne и PC равны.

Альтернативная гипотеза: средние пользовательские рейтинги платформ XOne и PC не равны.

```python
df_actuality.query('platform in ["XOne", "PC"]').dropna(subset=['user_score']).pivot_table(index='platform', values='user_score')
```

```python
x_one = df_actuality.query('platform == "XOne" & user_score != -1').dropna(subset=['user_score'])
pc = df_actuality.query('platform == "PC" & user_score != -1').dropna(subset=['user_score'])
```

```python
alpha = 0.05
result = st.ttest_ind(x_one['user_score'], pc['user_score'], equal_var=False)
print('p-значение:', result.pvalue)
if (result.pvalue < alpha):
    print('Отклоняем нулевую гипотезу')
else:
    print('Не отклоняем нулевую гипотезу')
```

Мы предположили, что средние пользвательские рейтинги платформ XOne и PC равны. Проверили гипотезу о равенстве средних двух генеральных совокупностей. 

Для этого посчитали p-value, используя ttest из библиотеки scipy для сравнения средних двух выборок. Использовали параметр equal_var=False, т.к. есть сомнения в равенстве дисперсий наших выборок.

Т.к. p-value больше выбранного уровня значимости, то нулевую гипотезу о равенстве средних мы не можем отвергнуть. Т.е. мы не можем утверждать, что средние пользовательские рейтинги платформ XOne и PC не равны. 

### Исследуем пользовательские рейтинги жанров Action и Sports

Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны.

Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports не равны.

```python
df_actuality.query('genre in ["Action", "Sports"]').dropna(subset=['user_score']).pivot_table(index='genre', values='user_score')
```

```python
action = df_actuality.query('genre == "Action" & user_score != -1').dropna(subset=['user_score'])
sports = df_actuality.query('genre == "Sports" & user_score != -1').dropna(subset=['user_score'])
```

```python
alpha = 0.05
result = st.ttest_ind(action['user_score'], sports['user_score'], equal_var=False)
print('p-значение:', result.pvalue)
if (result.pvalue < alpha):
    print('Отклоняем нулевую гипотезу')
else:
    print('Не отклоняем нулевую гипотезу')
```

Мы предположили, что средние пользвательские рейтинги жанров Action и Sports не отличаются. Проверили гипотезу о равенстве средних двух генеральных совокупностей. 

Для этого посчитали p-value, используя ttest из библиотеки scipy для сравнения средних двух выборок. Использовали параметр equal_var=False, т.к. есть сомнения в равенстве дисперсий наших выборок.

Т.к. p-value меньше выбранного уровня значимости, то нулевую гипотезу о равенстве средних отвергаем. Т.е. мы можем сказать, что средние пользовательские рейтинги жанров Action и Sports могут отличаться.
